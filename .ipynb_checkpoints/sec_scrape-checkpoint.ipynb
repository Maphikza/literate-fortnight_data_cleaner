```python
from pathlib import Path
import pandas as pd
import re
```

```python
with open(path) as file:
    text = file.read()
```

```python
raw_text_list = text.split("(")
```

```python
text = " ".join(raw_text_list)
```

```python
patterns = ["Common Stock, Shares Outstanding",
    "As of ",]
for pattern in patterns:
    match = re.search(pattern, text)
    if match:
        print("Found:", match.group(0))
    else:
        print("No match found for:", pattern)
```
